```python
# Importing the pathlib and csv library
from pathlib import Path
import csv

# Setting the file path
csvpath = Path ("PyBank_budget_data.csv")


#Initializing Variables
count_profit_loss = 0
total_profit_loss = 0
profit_loss_values = 0
difference = 0

# Creating Lists to hold data (dates and net_changes between rows)
dates = []
net_changes = []

# Opening the csv file as an object and in read mode 'r'
with open(csvpath, 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter = ",")
    
    # Reading the header row first
    csv_header = next(csvreader)

    # Reading the first row after header
    first_row = next(csvreader)
    count_profit_loss += 1
    total_profit_loss += int(first_row[1])
    profit_loss_values = int(first_row[1])
    
    #Going through each row of data after the header & first row to calculate the differences:
    for row in csvreader:
        # Reading dates from row 0
        dates.append(row[0])
        
        # Calculate the difference between rows, then add it to list of net_changes
        difference = int(row[1])-profit_loss_values
        net_changes.append(difference)
        profit_loss_values= int(row[1])
        
        
        #Total number of months/count
        count_profit_loss += 1
        

        #Total net amount of Profits/losses:
        total_profit_loss = total_profit_loss + int(row[1])

    #Greatest increase in profits:
    greatest_increase = max(net_changes)
    increase_index = net_changes.index(greatest_increase)
    increase_date = dates[increase_index]

    #Greatest decrease in profits: 
    greatest_decrease = min(net_changes)
    decrease_index = net_changes.index(greatest_decrease)
    decrease_date = dates[decrease_index]

    #Average change in Profits/Losses between months
    average_change = round(sum(net_changes)/len(net_changes),2)
    

# Printing the analysis:
print("Financial Analysis")
print("----------------------------")
print(f"Total Months: {(count_profit_loss)}")
print(f"Total: ${(total_profit_loss)}")
print(f"Average Change: ${average_change}")
print(f"Greatest Increase in Profits: {increase_date} (${(greatest_increase)})")
print(f"Greatest Decrease in Profits: {decrease_date} (${(greatest_decrease)})")
# print (net_changes)


# txt File Output:
out_filepath = Path("PyBank_output.txt")

with open(out_filepath, 'w') as file:
    file.write(f"Financial Analysis\n")
    file.write(f"----------------------------\n")
    file.write(f"Total Months: {count_profit_loss} \n")
    file.write(f"Total: ${total_profit_loss} \n")
    file.write(f"Average Change: ${average_change} \n")
    file.write(f"Greatest Increase in Profits: {increase_date} (${(greatest_increase)}) \n")
    file.write(f"Greatest Decrease in Profits: {decrease_date} (${(greatest_decrease)}) \n")
    
# csv File Output:
#header = ["Total Months","Total", "Average Change", "Greatest Increase", "Greatest Decrease"]
#metrics = [ count_profit_loss, total_profit_loss, avg_change, ((greatest_date), (greatest_increase)), ((worst_date), (greatest_decrease)) ]
#out_filepath = Path("/Users/simar/Jupyter-Workspace/1. Python/output.csv")

#with open(out_filepath, 'w') as csvfile:
    #csvwriter = csv.writer(csvfile, delimiter =',')
    #csvwriter.writerow(header)
    #csvwriter.writerow(metrics)
```
